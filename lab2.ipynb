# Image with circle

$$r^2 = x^2 + y^2$$

$$ r <= rr$$

> rr - radius of the circle

> `np.meshgrid()`

```python
np.arange(5)
```

```python
# array 4 rows, 5 cols
X,Y = np.meshgrid(np.arange(5),np.arange(4))
```

```python
print(X,Y,sep='\n\n')
```

## Array coordinates: `(rows, cols)`

## XY coordinates: `(x=cols, y=rows)`

---

## axis OX: (0,0),(1,0),(2,0), (3,0)

## axis OY: (0,0), (0,1), (0,2), (0,3)

```python
r = (X**2 + Y**2)**0.5
r
```

```python
print(f'{ar.shape}:  {ar.shape[0]} / {ar.shape[1]}')
```

```python
# image `ar`
ar1 = ar.copy()
rr = 100

# translate V[vx,vy]
vy = int(ar.shape[0]/2)
vx = int(ar.shape[1]/2)

X,Y = np.meshgrid(np.arange(ar.shape[0])-vx, np.arange(ar.shape[1])-vy)
r = (X**2 + Y**2)**0.5
```

```python
ar1[r<rr] = 0
plt.imshow(ar1);
```

```python
ar2 = ar.copy()

rr1 = 75
rr2 = 150

# ring conditions
ar2[(r>rr1) & (r<rr2)] = [255,255,0]

# inside conditions
ar2[(r<=rr1)] = [255,0,0]

plt.imshow(ar2);
```
